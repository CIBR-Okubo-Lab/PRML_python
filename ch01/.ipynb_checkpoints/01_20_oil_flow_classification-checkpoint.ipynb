# 1.4 The Curse of Dimensionality

Oil flow data from [here](https://inverseprobability.com/3PhaseData.html).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

## Load oil flow data

```python
data = np.loadtxt('../data/oil_flow/DataTrn.txt')
```

```python
labels = np.loadtxt('../data/oil_flow/DataTrnLbls.txt')
```

```python
x6 = data[:, 5]
x7 = data[:, 6]
c = labels.argmax(axis=1)
```

```python
c_all = np.unique(c)
color_list = ['r', 'g', 'b']
```

```python
plt.figure(figsize=(4.5, 4.5))
for i, c_i in enumerate(c_all):
    plt.scatter(x6[c==c_i], x7[c==c_i], color=color_list[i], facecolor='none', lw=1.2, s=50)
plt.plot(0.4, 1.3, 'kx')  # new observation
plt.xlim(0, 1)
plt.ylim(0, 2)
plt.xticks(np.arange(0, 1.25, 0.25))
plt.yticks(np.arange(0, 2.5, 0.5))
plt.xlabel(r'$x_6$', fontsize=14)
plt.ylabel(r'$x_7$', fontsize=14)
plt.show()
```

## Naive approach to classification

```python
N = 4
x_grid = np.linspace(0, 1, N + 1)
y_grid = np.linspace(0, 2, N + 1)
P = np.zeros((N, N), dtype=np.int64)
```

```python
for i in range(N):
    for j in range(N):
        idx = (x6 >= x_grid[i]) & (x6 < x_grid[i+1]) & (x7 >= y_grid[j]) & (x7 < y_grid[j+1])  # find out data points within a grid
        mode, _ = stats.mode(c[idx])  # find the class that has the largest number of points in the grid
        P[i, j] = np.array(mode)
```

```python
plt.figure(figsize=(4.5, 4.5))
plt.pcolormesh(x_grid, y_grid, P.T, cmap=matplotlib.colors.ListedColormap(['red', 'green', 'blue']), alpha=0.3)
plt.hlines(y=np.arange(0, 2.1, 0.5), xmin=0, xmax=1, color='k', lw=1)
plt.vlines(x=np.arange(0, 1.1, 0.25), ymin=0, ymax=2, color='k', lw=1)
for i, c_i in enumerate(c_all):
    plt.scatter(x6[c==c_i], x7[c==c_i], color=color_list[i], facecolor='none', lw=1.2, s=50)
plt.plot(0.4, 1.3, 'kx')  # new observation
plt.xlim(0, 1)
plt.ylim(0, 2)
plt.xticks(np.arange(0, 1.25, 0.25))
plt.yticks(np.arange(0, 2.5, 0.5))
plt.xlabel(r'$x_6$', fontsize=14)
plt.ylabel(r'$x_7$', fontsize=14)
plt.show()
```
