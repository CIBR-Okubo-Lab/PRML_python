Plot the fraction of the volume in the range $r=1-\epsilon$ to $r=1$ for various dimensions D (Figure 1.22).

```python
from ipywidgets import interact
import matplotlib.pyplot as plt
import numpy as np
```

```python
D_list = [1, 2, 5, 20]
```

```python
epsilon = np.linspace(0, 1, 100)
```

```python
text_pos = [(0.6, 0.55), (0.45, 0.65), (0.315, 0.8), (0.15, 0.92)]

plt.figure(figsize=(4, 4))
for D, pos in zip(D_list, text_pos):
    V = 1 - (1 - epsilon) ** D  # (1.76)
    plt.plot(epsilon, V, 'b')
    plt.text(*pos, f'D = {D}', ha='left')
    
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xticks(np.arange(0, 1.2, 0.2))
plt.yticks(np.arange(0, 1.2, 0.2))
plt.xlabel(r'$\epsilon$', fontsize=12)
plt.ylabel('volume fraction', fontsize=12)
plt.show()
```

```python
def plot_results(D=1):
    plt.figure(figsize=(4, 4))
    V = 1 - (1 - epsilon) ** D  # (1.76)
    plt.plot(epsilon, V, 'b')
    plt.xlim(0, 1)
    plt.ylim(0, 1.01)
    plt.xticks(np.arange(0, 1.2, 0.2))
    plt.yticks(np.arange(0, 1.2, 0.2))
    plt.xlabel(r'$\epsilon$', fontsize=12)
    plt.ylabel('volume fraction', fontsize=12)
    return
```

```python
interact(
    plot_results,
    D = (1, 50, 1),
);
```
