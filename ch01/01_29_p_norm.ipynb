# p-norm

Reproduce Figure 1.29.

```python
from ipywidgets import interact
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
q_list = [0.3, 1, 2, 10]
x = np.linspace(-2, 2, 100000)  # need to capture the points near zero for q=0.3
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 7))
ax = axs.flatten()
for i, q in enumerate(q_list):
    ax[i].plot(x, np.abs(x) ** q, 'r')
    ax[i].set_xlim(-2, 2)
    ax[i].set_ylim(0, 2)
    ax[i].set_xticks(np.arange(-2, 3, 1))
    ax[i].set_yticks(np.arange(0, 3, 1))
    ax[i].set_xlabel(r'$y - t$')
    ax[i].set_ylabel(r'$|y - t|^q$')
    ax[i].text(-0.5, 1.4, f'q = {q}')
    ax[i].tick_params(left=True, right=True, top=True, bottom=True, direction='in')
plt.show()
```

# Interactive plot

```python
def plot_results(q):
    
    x = np.linspace(-2, 2, 100000)
    
    fig, ax = plt.subplots(1, 1, figsize=(6, 4))
    ax.plot(x, np.abs(x) ** q, 'r')
    ax.set_xlim(-2, 2)
    ax.set_ylim(0, 2)
    ax.set_xticks(np.arange(-2, 3, 1))
    ax.set_yticks(np.arange(0, 3, 1))
    ax.set_xlabel(r'$y - t$')
    ax.set_ylabel(r'$|y - t|^q$')
    ax.tick_params(left=True, right=True, top=True, bottom=True, direction='in')
    plt.show()
```

```python
interact(
    plot_results,
    q = (0, 10, 0.1)
);
```
