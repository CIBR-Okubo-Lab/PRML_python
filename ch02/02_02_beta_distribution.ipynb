# beta distribution

Figure 2.2

```python
from ipywidgets import interact
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
```

```python
param_list = [(0.1, 0.1), (1, 1), (2, 3), (8, 4)]
x = np.linspace(0, 1, 1000)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 7))
ax = axs.flatten()
for i, (a, b) in enumerate(param_list):
    ax[i].plot(x, beta.pdf(x, a, b), 'r')
    ax[i].set_xlim(0, 1)
    ax[i].set_ylim(0, 3)
    ax[i].set_xticks(np.arange(0, 1.5, 0.5))
    ax[i].set_yticks(np.arange(0, 4, 1))
    ax[i].set_xlabel(r'$\mu$')
    ax[i].text(0.1, 2.6, f'a = {a}')
    ax[i].text(0.1, 2.3, f'b = {b}')
    ax[i].tick_params(left=True, right=True, top=True, bottom=True, direction='in')

plt.show()
```

# Interactive plot

```python
def plot_results(a, b):
    
    x = np.linspace(0, 1, 1000)
    
    fig, ax = plt.subplots(1, 1, figsize=(6, 4))
    ax.plot(x, beta.pdf(x, a, b), 'r')
    ax.plot(x, beta.pdf(x, a, b), 'r')
    ax.set_xlim(-0.01, 1.01)
    ax.set_ylim(0, 8)
    ax.set_xticks(np.arange(0, 1.5, 0.5))
    ax.set_yticks(np.arange(0, 10, 1))
    ax.set_xlabel(r'$\mu$')
    ax.tick_params(left=True, right=True, top=True, bottom=True, direction='in')
    plt.show()
```

```python
interact(
    plot_results,
    a = (0.1, 10, 0.1),
    b = (0.1, 10, 0.1)
);
```
