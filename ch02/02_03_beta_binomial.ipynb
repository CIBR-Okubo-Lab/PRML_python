Figure 2.3: one step update of a beta-binomial model

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta, binom
from scipy.special import comb
```

```python
a = 2
b = 2
```

```python
prior = beta(a, b)
posterior = beta(a + 1, b)
```

```python
mu = np.linspace(0, 1, 100)
```

```python
N = 1  # total number of trials
m = 1  # total number of heads
```

```python
likelihood = comb(N, m) * (mu ** m) * (1 - mu) ** (N - m)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 2))
ax[0].plot(mu, prior.pdf(mu), 'r')
ax[0].set_xlim(0, 1)
ax[0].set_ylim(0, 2)
ax[0].set_xticks([0, 0.5, 1])
ax[0].set_yticks([0, 1, 2])
ax[0].text(0.05, 1.7, 'prior', fontsize=12)
ax[0].set_xlabel(r'$\mu$', fontsize=12)
ax[0].tick_params(direction='in', which='both', right=True, top=True)

ax[1].plot(mu, likelihood, 'b')
ax[1].set_xlim(0, 1)
ax[1].set_ylim(0, 2)
ax[1].set_xticks([0, 0.5, 1])
ax[1].set_yticks([0, 1, 2])
ax[1].text(0.05, 1.7, 'likelihood function', fontsize=12)
ax[1].set_xlabel(r'$\mu$', fontsize=12)
ax[1].tick_params(direction='in', which='both', right=True, top=True)

ax[2].plot(mu, posterior.pdf(mu), 'r')
ax[2].set_xlim(0, 1)
ax[2].set_ylim(0, 2)
ax[2].set_xticks([0, 0.5, 1])
ax[2].set_yticks([0, 1, 2])
ax[2].text(0.05, 1.7, 'posterior', fontsize=12)
ax[2].set_xlabel(r'$\mu$', fontsize=12)
ax[2].tick_params(direction='in', which='both', right=True, top=True)
plt.show()
```
