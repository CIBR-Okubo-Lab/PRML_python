```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gamma
```

Be careful with the parameterization: https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.gamma.html

```python
x = np.linspace(0, 2, 200)
```

```python
def plot_gamma(a, b, ax):
    g = gamma(a, scale=1./b)  
    ax.plot(x, g.pdf(x), 'r')
    ax.text(1.2, 1.6, f'a = {a}', fontsize=12)
    ax.text(1.2, 1.4, f'b = {b}', fontsize=12)
    ax.set_xlim(0, 2)
    ax.set_ylim(0, 2)
    ax.set_xticks([0, 1, 2])
    ax.set_yticks([0, 1, 2])
    ax.tick_params(direction='in', which='both', right=True, top=True)
    ax.set_xlabel(r'$\lambda$')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 2.5))
plot_gamma(a=0.1, b=0.1, ax=ax[0])
plot_gamma(a=1, b=1, ax=ax[1])
plot_gamma(a=4, b=6, ax=ax[2])
```
