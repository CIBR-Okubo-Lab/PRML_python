```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t
```

```python
x = np.linspace(-5, 5, 200)
```

```python
p_0 = t.pdf(x, df=1e10)
p_1 = t.pdf(x, df=1.0)
p_2 = t.pdf(x, df=0.1)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, p_0, 'g', label=r'$\nu \rightarrow \infty$')
plt.plot(x, p_1, 'b', label=r'$\nu = 1.0$')
plt.plot(x, p_2, 'r', label=r'$\nu = 0.1$')
plt.legend(fontsize=12, frameon=False)
plt.xticks([-5, 0, 5])
plt.yticks(np.arange(0, 0.6, 0.1))
plt.xlim(-5, 5)
plt.ylim(0, 0.5)
plt.tick_params(labelsize=10, direction='in', which='both', right=True, top=True)
plt.show()
```
