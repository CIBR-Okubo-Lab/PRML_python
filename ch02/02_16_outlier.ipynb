Examine the effect of outliers on Gaussian (Figure 2.16)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

# Original data without outliers (N = 30)

```python
x = np.loadtxt('02_16_data.txt', skiprows=0)
```

```python
x.shape
```

```python
# ML estimate of a 1-D Gaussian
mu_0 = np.mean(x)
sigma_0 = np.std(x, ddof=0)
```

```python
mu_0, sigma_0
```

```python
x_grid = np.linspace(-5, 10, 100)
p_0 = norm(loc=mu_0, scale=np.sqrt(sigma_0)).pdf(x_grid)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 5))
ax.hist(x, bins=np.arange(-5.25, 10.5, 0.5), density=True, rwidth=0.7, edgecolor='black', facecolor='darkblue')
ax.plot(x_grid, p_0, 'g')
ax.set_xlim(-5, 10)
ax.set_ylim(0, 0.5)
ax.set_xticks(np.arange(-5, 15, 5))
ax.set_yticks(np.arange(0, 0.6, 0.1))
ax.tick_params(direction='in', which='both', right=True, top=True)
plt.show()
```

# Add outliers

```python
outliers = np.array([8.7, 9.23, 9.47])
```

```python
xx = np.concatenate((x, outliers))
```

```python
# ML estimate of a 1-D Gaussian
mu_1 = np.mean(xx)
sigma_1 = np.std(xx, ddof=0)
```

```python
mu_1, sigma_1
```

```python
p_1 = norm(loc=mu_1, scale=np.sqrt(sigma_1)).pdf(x_grid)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 5))
ax.hist(xx, bins=np.arange(-5.25, 10.5, 0.5), density=True, rwidth=0.7, edgecolor='black', facecolor='darkblue')
ax.plot(x_grid, p_1, 'g')
ax.set_xlim(-5, 10)
ax.set_ylim(0, 0.5)
ax.set_xticks(np.arange(-5, 15, 5))
ax.set_yticks(np.arange(0, 0.6, 0.1))
ax.tick_params(direction='in', which='both', right=True, top=True)
plt.show()
```
