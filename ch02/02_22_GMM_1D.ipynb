```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
N = 200  # number of points to sample from the x-axis
```

```python
pi = np.array([0.35, 0.3, 0.35])   # mixing coefficients
mu = np.array([-1.5, 0.0, 1.1])    # means
sigma = np.array([0.35, 0.55, 0.25])  # variances
```

```python
x = np.linspace(-3, 3, N)
```

```python
p = np.zeros((3, N), dtype=np.float64)
```

```python
for k in range(len(pi)):
    p[k] = pi[k] * norm.pdf(x, loc=mu[k], scale=sigma[k])
p_gmm = p.sum(axis=0)
```

```python
plt.figure(figsize=(6, 3))
for k in range(len(pi)):
    plt.plot(x, p[k] ,'b')
plt.plot(x, p_gmm, 'r')
plt.xticks([])
plt.yticks([])
plt.xlabel('x', fontsize=12)
plt.ylabel('p(x)', fontsize=12)
plt.show()
```
